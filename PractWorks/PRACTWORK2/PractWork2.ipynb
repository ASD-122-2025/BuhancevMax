## Практична робота №2. Асимптотична складність алгоритмів. Інші нотації
## Виконав: Буханцев Максим 

## Контрольні запитання:

**1. Що таке асимптотична складність алгоритму?**
Це оцінка ефективності алгоритму при великих об’ємах вхідних даних, що показує, як швидко зростає кількість операцій у залежності від розміру даних n."

**2. Які інші нотації, крім O-нотації, використовуються для вираження асимптотичної складності?**
Ω (Омега-нотація) Θ (Тета-нотація) o (Мала о-нотація) ω (Мала омега-нотація)

**3. Як визначити асимптотичну складність алгоритму за допомогою символів Θ і Ω?**
Θ(g(n)) — якщо f(n) ≈ g(n) з обох боків (знизу і зверху). Ω(g(n)) — якщо f(n) ≥ c·g(n) для великих n.

**4. Яка різниця між O-нотацією, Θ-нотацією і Ω-нотацією?**
O — максимум (верхня межа) Ω — мінімум (нижня межа) Θ — точне зростання (межі збігаються)

**5. Які основні властивості інших нотацій, таких як o (маленька о), ω (маленька омега) та 𝑜 (маленька о з верхнім індексом)?**
o(g(n)) — f(n) зростає повільніше за g(n); ω(g(n)) — f(n) зростає швидше за g(n); o⁽ⁿ⁾ — маленька о з верхнім індексом - позначення дуже швидкого спадання відносно іншої функції.
